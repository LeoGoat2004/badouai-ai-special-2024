```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 灰度图像直方图均衡化
img = cv2.imread("lenna.png", 1) # 1表示以彩色模式读取
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_dst = cv2.equalizeHist(img_gray)

# 直方图
hist = cv2.calcHist([img_dst],[0],None,[256],[0,256])
'''
[img_dst]：要计算直方图的图像。
[0]：表示计算图像的第一个通道（对灰度图像来说只有一个通道）。
None：没有掩膜。
[256]：直方图的 bins 数量，这里是 256，表示灰度级的数量。
[0, 256]：灰度级的范围。

'''
plt.figure(figsize=(12, 6))  # 宽度为6，高度为6
plt.subplot(121)
plt.hist(img_gray.ravel(), 256) # 将均衡化后的图像展平为一维数组，并计算直方图。256 是 bins 的数量。
plt.title('img_gray')
plt.subplot(122)
plt.hist(img_dst.ravel(), 256) 
plt.title('img_dst')
plt.show()

cv2.imshow("Histogram Equalization", np.hstack([img_gray, img_dst]))

cv2.waitKey(0)
```

```python
# 彩色图像直方图均衡化
img = cv2.imread("lenna.png", 1)

# 彩色图像均衡化,需要分解通道 对每一个通道均衡化
(b, g, r) = cv2.split(img)
bH = cv2.equalizeHist(b)
gH = cv2.equalizeHist(g)
rH = cv2.equalizeHist(r)
# 合并每一个通道
result = cv2.merge((bH, gH, rH))

cv2.imshow("Histogram Equalization", np.hstack([img, result]))

cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import numpy as np
```

```python
img_test = plt.imread("lenna.png")
img = plt.imread("Fricke.jpg")
```

```python
# lenna
# 原图
plt.subplot(221)
plt.imshow(img_test)
# 灰度化
img_test_gray = rgb2gray(img_test)
plt.subplot(222)
plt.imshow(img_test_gray,cmap='gray')
# 二值化
img_test_binary = np.where(img_test_gray >= 0.5, 1, 0) 
plt.subplot(223)
plt.imshow(img_test_binary,cmap='gray')
```

```python
# Fricke
# 原图
plt.subplot(221)
plt.imshow(img)
# 灰度化
img_gray = rgb2gray(img)
plt.subplot(222)
plt.imshow(img_gray,cmap='gray')
# 二值化
img_binary = np.where(img_gray >= 0.5, 1, 0) 
plt.subplot(223)
plt.imshow(img_binary,cmap='gray')
```
